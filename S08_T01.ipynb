## Exercici 1
Agafa un conjunt de dades de tema esportiu que t'agradi i normalitza els atributs categòrics en dummy. Estandaritza els atributs numèrics amb StandardScaler

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler


df_birds = pd.read_csv("Minnesota Birds.csv")

"""
Para simplificar la visualización elijo solo un atributo categórico, 
ya que en este dataset los categóricos tienen muchas categorías diferentes
"""

df_birds['Conservation status'] = df_birds["Conservation status"].fillna("Undefined")
df_birds = df_birds.fillna(0)
display(df_birds.info())

df_birds_dummy = pd.get_dummies(df_birds['Conservation status'])
display(df_birds_dummy)

df_birds_num = df_birds.loc[:,(df_birds.dtypes=='float').values]

df_birds_num_dummy = pd.concat([df_birds_num, df_birds_dummy], axis=1)

scaler = StandardScaler()

birds_withDummy_scaled = scaler.fit_transform(df_birds_num_dummy)
birds_withDummy_scaled = pd.DataFrame(birds_withDummy_scaled)
display(birds_withDummy_scaled)

birds_onlyNum_scaled = scaler.fit_transform(df_birds_num)
birds_onlyNum_scaled = pd.DataFrame(birds_onlyNum_scaled)
birds_onlyNum_scaled = pd.concat([birds_onlyNum_scaled, df_birds_dummy], axis=1)
display(birds_onlyNum_scaled)
```

Arriba he realizado una comparación del dataset con dummies escaladas y sin escalar. Como puede comprobarse, la estandarización de las dummies no plantea un cambio sustancial en los valores. Por ello, he optado por solo escalar las numéricas en el resto de ejercicios, y así conseguir que el cálculo sea computacionalmente más eficiente.
